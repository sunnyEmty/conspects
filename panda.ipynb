# Data Frame

**Data Frame** - набор двумерных данных с метками (набор с массивов Series, каждый массив соответствует своей метке)

Они создаются из:
1. Словаря одномерных массивов np
2. Двумерного массива np
3. из Структурированных массивов
4. из Series масссивов
5. Другого Data Frame

**Структурированный массив** - это массив, который организован как последовательность именованных полей (столбцов), причем каждое поле может содержать данные любого типа.

```python
users = np.array([
    ("Mike", 45, 1.5),
    ("Jack", 22, 2),
    ("Katya", 55, 1)],
dtype=[('name', 'U10'), ('age', 'i4'), ('hight', 'i4')])
users
```

При создании таблиц сначала заносятся ее строки, потом названия столбцов и количество памяти выделямой под каждый столбец

```python
# полем одного структурированного массива может являтся другой 
# массив
types = [('name', 'U10'), 
         ('age', 'i4'), 
         ('hight', 'i4'),
         ('Location', [('Country', 'U10'), ('City', 'U10')])]
users = np.array([
    ("Mike", 45, 1.5, ('UK', 'London')),
    ("Jack", 22, 2, ('USA', 'Detroit')),
    ("Katya", 55, 1, ('Russia', "Moscow"))],
    dtype=types)
users
```

combine выбирает какой столбец (одного из двух фреймов) добавить в соответствии результатом функции

```python
def takecol_with_sm_sum(col1, col2):
    return col1 if col1.sum() < col2.sum() else col2
# takecol выбирает колонку с мин. длиной

df1.combine(df2, takecol_with_sm_sum)
```

# Заполнение при переиндексации

функция reindex() использует метод, который является методом заполнения, выбранным из следующей таблицы:

```python
pd.DataFrame({
    'Метод': ['pad / ffill', 'bfill / backfill', 'nearest'],
    'Действие': []
})
```

```python
ind = pd.date_range('01.02.1909', periods=10)
s1 = pd.Series(np.random.random(10), index=ind)
s1
```

```python
s2 = s1['1909-01-02':'1909-01-07 ':2].copy()
s2
```

```python
s2.reindex(s1.index)
# для того чтобы элементы вошли в массив должно они должны быть
# одинаковы и в первом и во втором массиве
```

# Групперовка данных (groupby)

В данной концепции мы можем разбить данные на группы и применть функцию к каждой группе

Этот процесс обычно включает в себя несколько шагов:
1. Разделение: данные делятся на группы с применением некоторого условия
2. Примененне: процесс в котором мы применяем функцию к каждой группе 
3. Объединение: процесс в котором мы объединяем полученные результаты

Из указанной колонки (в квадратных скобочках) берутся элементы и разбиваются на группы в зависемости от меток в соответствующем категориальном столбце (сколько категорий стролько и групп)
пример
df.groupby('label')['hight']

```python
df = pd.DataFrame({
    'label': pd.Categorical(['A', 'B', 'A', 'D', 'D', 'A', 'B']),
    'hight': [134, 1, 32, 23, 3, 4, 5]
})
df
```

```python
df.info()
```

```python
df.groupby('label')['hight']
```

```python
df.groupby('label')['hight'].sum() # подсчитывается сумма элементов каждой группы столбца hight
```

```python
df.groupby('label')['hight'].mean() # cреднее
```

```python
df.groupby('label')['hight'].max()
```

```python
df.groupby('label')['hight'].unique()
```

```python
func = [np.min, np.max, np.mean, np.std]
df.groupby('label')['hight'].agg(func) # можно вызвать сразу несколько функций для каждой группы
```
